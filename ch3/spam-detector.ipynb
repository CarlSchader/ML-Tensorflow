```python
import email

DATA_DIR = '../data/spamandham/'
HAM_DIRS = ['easy_ham', 'easy_ham_2', 'easy_ham_3', 'hard_ham', 'hard_ham']
SPAM_DIRS = ['spam', 'spam_2', 'spam_3', 'spam_4']

file = open(DATA_DIR + '/easy_ham/2522.dc159c06c191406ece1b0fc98a352bbe')
some_email_string = file.read()
file.close()
print(some_email_string)
```

```python
some_email = email.message_from_string(some_email_string)
print(some_email.get_payload())
```

```python
import os
X_raw = []
y = []
for sub_dir in HAM_DIRS:
    for file_name in os.listdir(os.path.join(DATA_DIR, sub_dir)):
        file = open(os.path.join(DATA_DIR, sub_dir, file_name), errors='replace')
        X_raw.append(file.read())
        y.append(0)
        file.close()

for sub_dir in SPAM_DIRS:
    for file_name in os.listdir(os.path.join(DATA_DIR, sub_dir)):
        file = open(os.path.join(DATA_DIR, sub_dir, file_name), errors='replace')
        X_raw.append(file.read())
        y.append(1)
        file.close()
    
print(X_raw[0])
print(y[0])
```

```python
from sklearn.utils import shuffle

X_raw, y = shuffle(X_raw, y, random_state=0)
print(X_raw[0])
print(y[0])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_raw, y, test_size=0.2, random_state=0)
print(len(X_raw), len(y))
print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

```python
# construct word dictionary
pre_word_dict = {}
for email in X_raw:
    for word in email.split():
        if not(word in pre_word_dict):
            pre_word_dict[word] = 0
        else:
            pre_word_dict[word] += 1

word_dict = {}
word_index = 0
MIN_WORDS = 5
for word in pre_word_dict:
    if pre_word_dict[word] >= MIN_WORDS:
        word_dict[word] = word_index
        word_index += 1

print(len(pre_word_dict), len(word_dict))
```

```python
import numpy as np
def string_to_encoding(string, word_dict):
    vec = np.zeros(shape=(len(word_dict),), dtype=np.int32)
    for word in string.split():
        if word in word_dict:
            vec[word_dict[word]] += 1
    return vec

print(string_to_encoding(some_email_string, word_dict))
```

```python
# Encode data
X_train_encoded = np.array([string_to_encoding(email, word_dict) for email in X_train])
X_test_encoded = np.array([string_to_encoding(email, word_dict) for email in X_test])
print(X_train_encoded[0])
```

```python
with np.printoptions(threshold=np.inf):
    print(X_train_encoded[10])
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_predict

sgd_clf = SGDClassifier(random_state=0)
train_pred = cross_val_predict(sgd_clf, X_train_encoded, y_train, cv=3)
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib as mat
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

def plot_model_performance(y, pred):
    precisions, recalls, thresholds = precision_recall_curve(y, pred)
    plt.plot(precisions[1:], thresholds, 'b-', label='Precision')
    plt.plot(recalls[1:], thresholds, 'g-', label='Recall')
    plt.show()
    fpr, tpr, thresholds = roc_curve(y, pred)
    plt.plot(fpr, tpr, linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.show()

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_train, train_pred))
plot_model_performance(y_train, train_pred)
```

```python
clf = SGDClassifier(random_state=0)
clf.fit(X_train_encoded, y_train)
test_pred = clf.predict(X_test_encoded)
print(confusion_matrix(y_test, test_pred))
```
